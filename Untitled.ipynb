```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#loading the dataset and reading the head
raw_data =pd.read_csv('survey.csv')
raw_data.head()
```

```python
#Data Cleaning 
data=raw_data[raw_data['Gender']==('M')].index
raw_data.drop(data, inplace=True)
raw_data1=raw_data.drop(['comments'], axis=1)
raw_data1.head(100)
```

```python
treatment_yes=raw_data1[raw_data1['treatment'].str.contains("Yes")]
treatment_yes.head()
```

```python
mentalhealth_by_location =treatment_yes.groupby(['Country'])['treatment'].count().reset_index(name="of_Employees")
mentalhealth_by_location.sort_values("of_Employees", axis=0, ascending = False)
mentalhealth_by_location.head(100)
```
